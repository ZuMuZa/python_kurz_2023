```python
u202 = pandas.read_csv("u202.csv")
u203 = pandas.read_csv("u203.csv")
u302 = pandas.read_csv("u302.csv")
u202
```

## Chybějící hodnoty

```python
u202.info()
```

```python
u202["znamka"].isna() #stejně tak .isnull()
```

kterým studentům chybí známka ve tříde u202

```python
u202[u202["znamka"].isna()] #.isnull()
```

```python
u202.dropna(subset=["znamka"]) #zkontroluj u sloupcu me zajímají (znamka) kde je nulová hodnota a ty data odstran
# do subset zadávám ty sloupce, které chci aby se zkontrolovali NaN hodnoty, bez subset se zkontrolují všechny sloupce 
# #u202.dropna(subset=[]) nepodívá se na žádný sloupec!
u202.dropna()
```

## Spojení tabulek

```python
u202["mistnost"] = "u202"
u203["mistnost"] = "u203"
u302["mistnost"] = "u302"
u202
```

```python
#pandas.concat([u202, u203, u302]) index se opakuje
maturita = pandas.concat([u202, u203, u302], ignore_index=True)
maturita_pritomni_studenti = maturita.dropna(subset=["znamka"])
maturita_pritomni_studenti
```

```python
maturita.info()
maturita.to_csv("maturita.csv",index=False)
```
