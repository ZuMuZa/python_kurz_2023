Data si stáhni [zde](https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/nacteni-dat/staty.json). Jsou ve formátu JSON, proto použijeme vhodnou metodu `read_json`:

```python
staty = pandas.read_json('staty.json')
staty
staty.info()
```

Ukážeme si, jak filtrovat řádky v tabulce na základě hodnot ve sloupcích. Například, které státy mají počet obyvatel menší než 10000?

```python
staty["population"]
```

```python

#print(staty["population"] < 10_000)
```

```python
staty[staty["population"] < 10_000]
```

Které státy leží v subregionu `"South America"`?

```python
staty_jizni_ameriky = staty[staty["subregion"] == "South America"]
staty_jizni_ameriky["name"]
```

Které státy mají nízký index ekonomické nerovnosti, např. menší než 30?

```python
staty[staty["gini"] < 30] # indexy se zachovávají
staty[staty["gini"] < 30].reset_index() #index zustane a přidá se nový číslovaný od začátku
staty[staty["gini"] < 30].reset_index(drop=True) #jen nový číslovaný od začátku
#index_gini_nizky = staty[staty["gini"] < 30]
#index_gini_nizky["name"]
```

## Kombinace podmínek
Které státy jsou velké rozlohou (např. víc než 100 000), a mají zároveň malý počet obyvatel (méně než 1 milion)?

```python
#staty[staty["area"] > 100_000][staty["population"] < 1_000_000]

staty[(staty["population"] < 1_000_000) & (staty["area"] > 100_000)]
```

Které státy leží v Evropě nebo v Africe?

```python
#staty[staty["region"] == "Europe"] | [staty["region"] == "Africa"] | [...] | [...] # lze řetězit podmínky ale je to nepřehledné->ISIN
staty[(staty["region"] == "Europe") | (staty["region"] == "Africa")] # s kulatými závorka je to přehlednější ale lze i bez nich
```

## Metoda `isin`
Které státy leží v Evropě nebo v Africe?

```python
staty[staty["region"].isin(["Europe", "Africa"])]
```

## Metoda `between`
Které státy mají počet obyvatel mezi jedním a pěti miliony?

```python
#staty[(staty["population"] > 1_000_000) & (staty["population"] < 5_000_000)]
staty[staty["population"].between(1_000_000, 5_000_000, inclusive='both')]
```
